```python
"""
    larbi.saifi@gmail.com
    5 joueurs
    2000 matchs
    pierre, feuille, ciseaux
    
"""

import random # imposter la bibliothèque qui vas nous permettre de générer les chiffres aléatoires

# liste des joueurs avec pour chacun d'eux:
# son nom: représente la clé du dictionnaire ( cherche sur google "dictionnaire python" )
# son score (= 0 au début),
# la proba de tirer feuille, ciseaux, pierre ( ex: Ema vas tirrer dans 20% des cas feuille, 50% des cas ciseaux, et le reste du temps pierre )

joueurs = { 
            "Ema":     {"score": 0, "proba": [0.2,0.5,0.3]}, 
            "Patrick": {"score": 0, "proba": [0.3,0.4,0.3]},
            "Lucas":   {"score": 0, "proba": [0.1,0.3,0.6]},
            "Fred":    {"score": 0, "proba": [0,0.5,0.5]},
            "Suzanne": {"score": 0, "proba": [0.9,0,0.1]}
           }

# la fonction choix coup, prend comme argument le nom du joueur
# elle renvoie soit l'indice 1, 2, 3 ( feuille, ciseaux, pierre ) selon la distribution de proba de chaque joueur
def choix_coup(nom_joueur):
    r = random.random() # générer un chiffre aléatoire entre 0 et 1
    probs = joueurs[nom_joueur]["proba"] # récupérer les proba du joueur
    
    # On cherche à quel classe appartient le chiffre r qu'on viens de tirer
    # lire la formule https://fr.wikipedia.org/wiki/Loi_de_probabilité#D.C3.A9finition_math.C3.A9matique
    index = 1
    while(r >= 0 and index < len(probs)):
      r -= probs[index]
      index += 1
    return index - 1


nb_matchs = 2000 # nombre de matchs
jeu = {1: "feuille", 2: "ciseaux", 3: "pierre"} # dictionnaire "puissance" : "coup" ordonné par puissance ( sauf exception de 1 > 3 ) 

for match in range( nb_matchs ):
    
    # choisir deux joueurs DIFFERENTS
    while True:
        j1 = random.choice( joueurs.keys() )
        j2 = random.choice( joueurs.keys() )
        if j1 != j2:
            break;
    
    # choix d'une main pour chaque joueur
    # utiliser notre fonction choix_coup() qui utilise les proba de chaque joueur
    jeu_j1 = jeu[choix_coup(j1)] 
    jeu_j2 = jeu[choix_coup(j2)]
    
    
    if (jeu_j1 != jeu_j2 ): # si match nule, ne pas compter le point
        
        # si (joueur 1 == feuille ET joueur 2 == pierre ) 
        # sinon si ( joueur 1 "est plus puissant que" joueur 2 )
        # c'est a dire si (joueur 1 == ciseaux ET joueur 2 == feuille ) 
        # ou si (joueur 1 == pierre ET joueur 2 == ciseau ) 
        if (jeu_j1 == 1 and jeu_j2 == 3) or (jeu_j1 > jeu_j2): 
            joueurs[j1]["score"] += 1 # alors joueur 1 vas avoir le point
        else:
            joueurs[j2]["score"] += 1 # sinon dans tous les autres cas, à part ceux cité en haut, c'est joueur 2 qui aura le point


# mettre les résultats sous forme de dictionnaire {nom_joueur: score}
resultats = {k:v["score"] for k,v in joueurs.items() }

# ordonner les résultats par score
resultats_ordonnees =  sorted(resultats.items(), key=lambda x:x[1], reverse=True)

# Ecrire dans l'ordre le classement des joueurs ( avec leurs score )
for i in range( len(resultats_ordonnees) ):
    print "#%d %s (score: %d)" % (i+1, resultats_ordonnees[i][0], resultats_ordonnees[i][1])

```
